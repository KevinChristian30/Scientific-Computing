### Cubic Interpolation

Interpolation adalah analisis matematika yang digunakan untuk melakukan estimasi sebuah titik baru dengan menggunakan data yang sudah diketahui. Untuk Cubic Interpolation, kita akan menggunakan function Cubic Spline. Cubic Spline berfungsi untuk menentukan titik Y yang didapatkan berdasarkan pembagian titik X

#### Code

```python
X = [0, 1, 2]
Y = [1, 3, 2]
```

```python
from scipy.interpolate import CubicSpline 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cubic_interpolation(x, y, segments = 100):
    f = CubicSpline(x, y)
    
    x_new = np.linspace(0, 2, segments)
    y_new = f(x_new)
    
    plt.plot(x, y, 'ro')
    plt.plot(x_new, y_new)
    plt.title('Cubic Spline Interpolation')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    plt.show()
```

```python
cubic_interpolation(X, Y)
```
