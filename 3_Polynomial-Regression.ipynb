### Polynomial Regression

Polynomial Regression bertujuan untuk mencari persamaan garis yang paling cocok dengan data, dimana jika diberikan banyak titik pada sebuah graph, kita dapat membuat garis dimana jarak antara setiap titik dengan garis itu minimum

#### Code

```python
X = [1, 2, 3, 4, 5, 6, 7, 8]
Y = [2, 5, 1, 7, 9, 3, 6, 10]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polynomial_regression(x, y):
    x = np.array(x)
    y = np.array(y)
    
    plt.figure(figsize=(12, 8))
    for i in range(1, 7):
        plt.subplot(2, 3, i)
        plt.plot(x, y, 'o')
        
        y_est = np.polyfit(x, y, i)
        plt.plot(x, np.polyval(y_est, x))
        plt.title(f'Polynomial Order{i}')
```

```python
polynomial_regression(X, Y)
```
